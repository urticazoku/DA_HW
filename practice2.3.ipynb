```python
import re
n=open("text.txt",'r')
t=n.read()
l=re.split('[^a-z]', t)
l = [x for x in l if x!='']
n.close()
```

```python
dct = {}
count=0
for word in l:
    if word not in dct:
        dct[word] = count
        count+=1

d = len(dct)
nn = num_sent
print dct
print "###############"
dct=sorted(dct.items(), key=lambda x: x[1]) 
print dct
```

```python
n_dct=[]
for i in range(len(dct)):
    n_dct.append(dct[i][0])
n_dct
```

```python
import numpy as np
n=open("text.txt",'r')
my_list=[]
for sent in n:
    l=re.split('[^a-z]', sent)
    l = [x for x in l if x!='']
    my_list.append(l)
n.close()

words=[]
count=0
for sent in my_list:
    for n_word in n_dct:
        for word in sent:
            if word==n_word:
                count+=1
        words.append(count)
        count=0
words
```

```python
np_array=np.array(words).reshape((nn,d))
```
